```python
# 4- define function Routing_Table() (4 marks)
# this function takes its inputs from the user to build its routing table
# the routing table should look like the following :
#-------------------------------------------------
#    Network           |   output port number    |
#-------------------------------------------------
#  192.160.50.0/24    |    1                     |
#  192.100.50.0/24    |    2                     |
#  110.160.50.0/30    |    3                     |
#  150.80.0.0/16      |    4                     |
#  else               |    0                     |
#-------------------------------------------------
# the function returns  the routing table as a list of networks and relatve ouput ports interfaces
# the function should ask the user for the number of required interfaces first 
# then it asks the user to fill the routing table entries accordingly 
# if the user entered the character '0' as an egress interface then the function set this as default interface 
# in all cases the routing table should contain a default ouptut interface in case of any network address 
# arrive at the router while it is out of the range of its routing entries
#hint you can use list of dictionaries 
# Your Code Starts Here 
def Routing_Table():
  routing_table = {}
  # check valid integer interfaces
  while True:
    interfaces_count = input('Enter the number of interfaces:')
    if interfaces_count.isnumeric() and  int(interfaces_count) >= 0:
      interfaces_count = int(interfaces_count)  
      break
   
  # get routing table entries from user
  while interfaces_count != 0:
    # get network address
    try:
      network = ipaddress.ip_network(input('Enter network address:'))
    except:
      print('invalid network address entered.')
      continue
    
    # get port number
    interface = input('Enter port number:')
    if interface in routing_table:
      print('port already exists.')
      continue
    
    # add to routing table
    routing_table[interface] = network
    interfaces_count = int(interfaces_count) - 1
  
  # add default gateway if not exists
  if '0' not in routing_table:
    routing_table['0'] = ipaddress.ip_network('255.255.255.255/32')
  
  return routing_table
    
```

```python
# 5- define function Forward (4 marks)
# the function takes as input the routing table as returend from function Routing_Table() and an ip address as string.
# the function should perform the longest prefix match between the ip and the reouting table
# then the function should decide (return and print) the suitable interface to forward the packet with the input ip to.
# Your Code Starts Here 

#the function output should like like the following:
#the interface  192.168.0.1 will be forwarded through port  1 

def Forward(routing_table, ip):
  possible_interfaces = []
  for interface, network in routing_table.items():
    if network.overlaps(ipaddress.ip_network(ip)):
      possible_interfaces.append((network.prefixlen, interface))
  if len(possible_interfaces) != 0:
    print('the interface', ip, 'will be forwarded through port', max(possible_interfaces)[1])
  else:
    print('the interface', ip, 'will be forwarded through port', '0')
```
